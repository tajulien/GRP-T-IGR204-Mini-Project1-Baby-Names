```python
import altair as alt 
import pandas as pd

alt.data_transformers.enable('json')
df_names = pd.read_csv('dpt2020.csv', sep=";")
```

```python
print(df_names.head(10))
print(df_names.shape)
```

```python
df_names = df_names[df_names.preusuel != "_PRENOMS_RARES"]
df_names = df_names[df_names.annais != "XXXX"]
```

```python
print(df_names.head(10))
print(df_names.isna().sum())
```

```python
df_names = df_names.drop('dpt', axis=1)
df_names.info()
```

```python
df_names.sort_values(by=['annais', 'preusuel'])
df_names['sexe'] = df_names["sexe"].map({1:"M",2:"F"})
df_names.sort_values(by=['annais', 'preusuel'])
```

```python
df_names['merge_unique'] = (df_names['sexe']+df_names['preusuel']+df_names['annais'])
df_names = df_names.sort_values(by=['annais', 'preusuel'])
df_names.reset_index(inplace = True)
```

```python
print(df_names.head(10))
```

```python
name_to_keep = []
unique = ''
for index, row in df_names.iterrows():
    if index == 0:
        unique = df_names.loc[index, 'merge_unique']
    else:
        unique = df_names.loc[index-1, 'merge_unique']
        if df_names.loc[index, 'merge_unique'] == unique:
            pass
        else:
            if str(df_names.loc[index, 'merge_unique'])[1:] == str(unique)[1:]:
                name_to_keep.append(str(df_names.loc[index, 'merge_unique']))
print(name_to_keep)
```

```python
name_to_keep2 = [x[1:] for x in name_to_keep]
```

```python
df2 = df_names.groupby(['annais', 'merge_unique','preusuel'])['nombre'].agg(['sum','count'])
```

```python
print(df2.head(20))
```

```python
df2.reset_index(inplace = True)
```

```python
df2["sexe"] = df2["merge_unique"].str[:1]
df2["merge_unique"] = df2["merge_unique"].map(lambda x: str(x)[1:])
df2 = df2[df2.merge_unique.isin(name_to_keep2)]
df2 = df2.sort_values(by=['annais', 'preusuel'])
df3 = df2.groupby(['annais'])['sum'].agg(['sum'])
```

```python
df3.reset_index(inplace = True)
```

```python
annee_selection = alt.selection_interval(encodings=['x'])

chart_1 = alt.Chart(df3, title='Évolution temporelle des prénoms mixtes').mark_line().encode(
    x=alt.X('annais:T', axis=alt.Axis(title='Année')), 
    y=alt.Y('sum:Q', axis=alt.Axis(title='Nombre de naissances avec prénom mixte')),
    tooltip=['annais:T','sum:Q']).properties(width=700).add_selection(annee_selection)
    

legend_1 = alt.Chart(df3).mark_point().encode(
    y=alt.Y('sum:Q', axis=alt.Axis(orient='right', title='Répartition des naissances')),
)
```

```python
(chart_1 | legend_1)
```

_Are there gender effects in the data? Does popularity of names given to both sexes evolve consistently? (Note: this data set treats sex as binary; this is a simplification that carries into this assignment but does not generally hold.)_

Comme nous le montre cette courbe temporelle, la popularité des prénoms mixtes a eu une grande évolution dans les premières 50 années de 1900. Ensuite une brusque décroissance vers la fin des années 1960 jusqu'à 1980 pour se stabiliser et croitre légèrement.<br>
Nous ne pouvons pas dire qu'il s'agisse d'une évolution constante à travers la représentation graphique associée.
